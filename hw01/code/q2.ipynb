```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    "Size": [1000, 5000, 10000, 20000, 50000],
    "InsertionSort (ms)": [2.21681, 56.805, 230.205, 913.366, 5736.25],
    "MergeSort (ms)": [1.41739, 7.5942, 15.8429, 32.2628, 85.8542],
    "BucketSort (ms)": [0.386627, 1.53927, 2.74471, 5.57126, 12.306]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df["Size"], df["InsertionSort (ms)"], marker='o', label='InsertionSort (ms)')
plt.plot(df["Size"], df["MergeSort (ms)"], marker='o', label='MergeSort (ms)')
plt.plot(df["Size"], df["BucketSort (ms)"], marker='o', label='BucketSort (ms)')

# Adding labels and title
plt.xlabel('Size of Input Array')
plt.ylabel('Time (ms)')
plt.title('Running Time Comparison of Sorting Algorithms')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
