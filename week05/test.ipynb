### Union of Segments

```python
import random

n = 1000
a = sorted(random.sample(range(1, 10000), n))
q = 1000
queries = []
segments = []

for _ in range(q):
    op = random.choice(['insert', 'delete', 'report'])
    i = random.randint(1, n - 1)
    j = random.randint(i + 1, n)
    if op == 'insert':
        segments.append((i, j))
        queries.append(f'insert {i} {j}')
    elif op == 'delete':
        if segments:
            idx = random.randint(0, len(segments) - 1)
            i, j = segments.pop(idx)
            queries.append(f'delete {i} {j}')
        else:
            # If no segments to delete, default to insert
            segments.append((i, j))
            queries.append(f'insert {i} {j}')
    else:
        queries.append(f'report {i} {j}')

print(n)
print(' '.join(map(str, a)))
print(q)
for query in queries:
    print(query)
```
